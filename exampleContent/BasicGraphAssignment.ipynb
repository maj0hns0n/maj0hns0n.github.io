### Do opioid overdoes tend to be associated with less affluent areas—that is, areas where families have lower incomes?

Yes, there appears to be a negative relationship between families with lower incomes and higher opioid overdoses. I chose a scatterplot to examine the relationship between the two numeric variables (MedianHouseholdIncome and FPOO-Rate) since it clearly identifies a cluster of data points in the lower-end of the income scale associated with higher overdose rates. 

```python
fig, ax = plt.subplots()

ax.scatter(df_op['MedianHouseholdIncome'], df_op['FPOO-Rate'], alpha=0.5, c = 'g')  
fig.suptitle('Median Income vs. Opioid Overdose Rate ')   
ax.xaxis.set_label_text('Median Household Income')  
ax.yaxis.set_label_text('Opioid Overdose Rate')       
fig.set_size_inches(7,5)
plt.show()
```

### Some people who start with opioid addictions are reported to transition to heroin use. What is the relationship in Virginia counties between opioid overdoses and heroin overdoses? 

The scatterplot of opioid overdose vs herion overdose rates does appear to present a pattern of higher heroin overdoses in the lower range of opioid overdoses which could be evidence to support a hypothesis of opioid acting as a gateway drug to heroin.

```python
fig, ax = plt.subplots()

ax.scatter(df_op['FPOO-Rate'], df_op['FFHO-Rate'], alpha=0.5, c = 'r')  
fig.suptitle('Opioid Overdose Rate vs. Heroin Overdose Rate ')   
ax.xaxis.set_label_text('Opioid Overdose Rate')  
ax.yaxis.set_label_text('Heroin Overdose Rate')       
fig.set_size_inches(7,5)
plt.show()
```

### How have vehicle sales in the United States varied over time?

The U.S. vehicle sales trend appears to be cyclical and perhaps closely follow overall economic trends in the U.S. I chose a line chart for this time series data for its ease in visualizing the data trend overtime.

```python
from pandas import read_csv
from matplotlib import pyplot
series = read_csv('TOTALNSA.csv', header=0, index_col=0, parse_dates=True, squeeze=True)
series.plot()
plt.title('US Vehicle Sales')
plt.xlabel('Year')
plt.ylabel('Total Sales')
pyplot.show()
```
